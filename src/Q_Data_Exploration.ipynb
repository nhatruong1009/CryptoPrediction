Giải nghĩa các cột dữ liệu:
* symbol : mã kí hiệu loại giao dịch
* base : mã đồng cơ sở
* quote : mã đồng báo giá
* Kline open/close time : thời gian mở/đóng của việc thu dữ liệu
* open price : Giá tại thời điểm đầu khi thu dữ liệu
* close price : giá tại thời điểm cuối cùng khi thu dữ liệu
* high price : giá cao nhất trong thời gian thu dữ liệu
* low price : giá thấp nhất trong thời gian thu dữ liệu
* volume : số lượng base đã được trao đổi
* quote asset volume : số lượng quote đã được trao đổi
* taker buy base volume : số lượng mua base (lệnh) (?)
* taker buy quote volume : số lượng mua quote (lệnh) (?)

```python
# print("The following inverse pairs exist:")
# for index, row in symbols.iterrows():
#     for index2, row2 in symbols.iterrows():
#         if index != index2 and row['baseAsset'] == row2['quoteAsset'] and row['quoteAsset'] == row2['baseAsset']:
#             print(f"{row['symbol']} and {row2['symbol']}")
duplicate_symbols = []
def func(symbol_row):
    matches = symbols.apply(lambda row: symbol_row["baseAsset"] == row["quoteAsset"] and row["baseAsset"] == symbol_row["quoteAsset"], axis=1)
    if matches.any():
        duplicate_symbols.append(symbol_row["symbol"])
        return symbol_row
    return "Fail"

symbols.apply(lambda symbol_row: func(symbol_row), axis=1)
```

```python
uni = pandas.concat([symbols['baseAsset'],symbols['quoteAsset']]).unique() # co 525 loai coin
arr = numpy.concatenate([symbols['baseAsset'],symbols['quoteAsset']])
count = dict()

for i in uni:
    count[i] = numpy.count_nonzero(arr == i)
pandas.Series({k: v for k, v in sorted(count.items(), key=lambda item: item[1],reverse=True)}).head(10)
# 5 loại đồng tiền đầu tiên xuất hiện cực kì nhiều, có thể đây là 5 loại đồng tiền nền của sàn (xuất hiện ở khoảng trên 50% loại đồng, tới khoảng 90%)
# bằng một số tra cứu, BUSD thuộc về đông tiền ổn định, và USDT cố định ở mức 1$, nên ta có thể coi như 3 đồng này là đồng nền chính của sàn
# BTC, ETH, BNB tuy xuất hiện nhiều nhưng giá không ổn định trên thị trường khi quy về tiền thật, nên không thể làm mốc để so sánh được
```

```python
base_coin = ['BUSD','USDT']
```

```python
f_symbols =  symbols[symbols.baseAsset.isin(base_coin) | symbols.quoteAsset.isin(base_coin)]
f_symbols = f_symbols[~(f_symbols.baseAsset.isin(base_coin) & f_symbols.quoteAsset.isin(base_coin))]
f_symbols
```

```python
f_data = data[data.Symbol.isin(f_symbols.symbol)]
f_data
```

```python
f_data[~f_data.quote.isin(base_coin)] # Các giá trị quy đổi cần điều chỉnh lại
```

```python
# kiểm tra số đồng có trong bảng
print( 'So luong coin co trong data: ', pandas.concat([f_data.base,f_data.quote]).nunique()) # số lượng thấp hơn khi tính ở trên, đó là do một số đồng ko còn hiện hữu nữa
print( 'So luong coin con giao dich: ', pandas.concat([symbols[symbols.status == 'TRADING'].baseAsset ,symbols[symbols.status == 'TRADING'].quoteAsset]).nunique())
```
