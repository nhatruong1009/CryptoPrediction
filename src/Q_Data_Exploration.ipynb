```python
import pandas
from pandas import DataFrame
import numpy
```

### Giải nghĩa các cột dữ liệu:
* symbol : mã kí hiệu loại giao dịch
* base : mã đồng cơ sở
* quote : mã đồng báo giá
* Kline open/close time : thời gian mở/đóng của việc thu dữ liệu
* open price : Giá tại thời điểm đầu khi thu dữ liệu
* close price : giá tại thời điểm cuối cùng khi thu dữ liệu
* high price : giá cao nhất trong thời gian thu dữ liệu
* low price : giá thấp nhất trong thời gian thu dữ liệu
* volume : số lượng base đã được trao đổi
* quote asset volume : số lượng quote đã được trao đổi
* taker buy base volume : số lượng mua base (lệnh) (?)
* taker buy quote volume : số lượng mua quote (lệnh) (?)

### Kiểm tra giá trị của các token 

làm theo các bước:
- kiếm ra các đồng tiền xuất hiện nhiều nhất ở các giao dịch
- tra cứu thông tin về các đồng tiền đó
- lọc lại dữ liệu theo quy đổi về đồng tiền nền (coi như đồng này không biến động)

```python
uni = pandas.concat([symbols['baseAsset'],symbols['quoteAsset']]).unique() # co 525 loai coin
arr = numpy.concatenate([symbols['baseAsset'],symbols['quoteAsset']])
count = dict()

for i in uni:
    count[i] = numpy.count_nonzero(arr == i)
pandas.Series({k: v for k, v in sorted(count.items(), key=lambda item: item[1],reverse=True)}).head(10)
```

- 5 loại đồng tiền đầu tiên xuất hiện cực kì nhiều, có thể đây là 5 loại đồng tiền nền của sàn (xuất hiện ở khoảng trên 50% loại đồng, tới khoảng 90%)
- bằng một số tra cứu, BUSD và USDT thuộc về đông tiền ổn định ở mức 1$, nên ta có thể coi như 2 đồng này là đồng nền chính của sàn
- BTC, ETH, BNB tuy xuất hiện nhiều nhưng giá không ổn định trên thị trường, nên không thể làm mốc để so sánh được

```python
base_coin = ['BUSD','USDT'] # đồng tiền ổn định giá
```

```python
f_symbols =  symbols[symbols.baseAsset.isin(base_coin) | symbols.quoteAsset.isin(base_coin)]
f_symbols = f_symbols[~(f_symbols.baseAsset.isin(base_coin) & f_symbols.quoteAsset.isin(base_coin))]
f_symbols.head()
```

```python
f_data = data[data.Symbol.isin(f_symbols.symbol)]
f_data.head()
```

```python
f_data[~f_data.quote.isin(base_coin)].head() # Các giá trị quy đổi cần điều chỉnh lại
```

```python
# kiểm tra số đồng có trong bảng
print( 'number of token in data: ', pandas.concat([f_data.base,f_data.quote]).nunique()) # số lượng thấp hơn khi tính ở trên, đó là do một số đồng ko còn hiện hữu nữa
print( 'number of token still trading: ', pandas.concat([symbols[symbols.status == 'TRADING'].baseAsset ,symbols[symbols.status == 'TRADING'].quoteAsset]).nunique())
```

biến bảng trên thành giá trị quy đổi qua đồng tiền nền (Thực tế đã là bảng hiện tại) <br>
tạo 1 bảng giá tiền của đồng tiền theo symbols <br>
Cần đổi lại các đồng tiền có quote không nằm trong 'base_coin' theo quy tắc:
- base <-> quote
- Volume <-> Quote asset volume
- Taker buy base asset volume <-> Taker buy quote asset volume
- Các giá trị tiền tệ: New = 1 / Old

```python
#only run this cell one
temp = f_data[~f_data.quote.isin(base_coin)]
f_data[~f_data.quote.isin(base_coin)] = pandas.concat([
    temp[['Symbol','quote','base' ,'Kline open time']],
    1/temp[['Open price','High price','Low price','Close price']],
    temp[['Quote asset volume','Kline Close time','Volume','Number of trades','Taker buy quote asset volume','Taker buy base asset volume','Unused field, ignore']]
],axis=1)

# f_data now is the data of "true value" of the token
all(f_data.quote.isin(base_coin))
```

```python
f_data.to_csv("Data/token_price.csv",index=False)
```
