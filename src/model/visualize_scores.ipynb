```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../datastore/model_scores/grid_search_cv_5_len_2500.csv")
```

# ðŸ”¥ Test score

```python

df.loc[df["mean_test_score"] > 0].plot(kind="bar", x="param_reg__estimator", y="mean_test_score", figsize=(15, 10))
fig = plt.figure() 
```

# ðŸ”¥ Compare train and predict time

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 6))

ax1 = df.plot(kind="bar", x="rank_test_score", y="mean_fit_time", ax=axes[0])
ax1.set_ylabel("train time (s)")
ax2 = df.plot(kind="bar", x="rank_test_score", y="mean_score_time", ax=axes[1])
ax2.set_ylabel("predict time (s)")
```
