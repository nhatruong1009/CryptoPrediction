```python
import pandas
import numpy as np
from pandas import DataFrame
```

```python
symbols = pandas.read_csv('symbols.csv').drop(['Unnamed: 0'], axis=1)
token = pandas.read_csv('token_price.csv')
data_symbol = pandas.read_csv('data.csv').drop(['Unnamed: 0'], axis=1)
```

```python
token
```

```python
data = token[token['Symbol'] == 'BTCUSDT'].set_index('Kline open time').sort_index()
```

```python
price = data[['High price','Low price']] #High price và low price giá tương quan trên toàn phiên giao dịch
price
```

```python
up_down = price[1:].reset_index() - price[:-1].reset_index()
up_down
```

```python
change_period_high = [1]
if (up_down['High price'][0] > 0):
    current = 1
else:
    current = -1
for i in up_down['High price'][1:]:
    if (i * current > 0):
        change_period_high[-1] += 1
    else: 
        change_period_high += [1]
```

```python
np.mean(change_period_high)
```

```python
change_period_low = [1]
if (up_down['Low price'][0] > 0):
    current = 1
else:
    current = -1
for i in up_down['Low price'][1:]:
    if (i * current > 0):
        change_period_low[-1] += 1
    else: 
        change_period_low += [1]
```

```python
np.mean(change_period_low)
```

#### Giá của các symbol có ảnh hưởng lẫn nhau không

```python
data_symbol.set_index('Kline open time')
```

```python
symbols_get = symbols[symbols['status'] == 'TRADING']
```

```python

other =  pandas.concat([data_symbol[data_symbol['Symbol']==i][['Kline open time','High price']].set_index('Kline open time') for i in symbols_get['symbol']],axis = 1)
other.columns = list(symbols_get['symbol'])
other = other.fillna(0)
other = other.astype(float)
```

```python
other
```

```python
other.corr()
```

```python
corralation = other.corr()
```

```python
top = dict(zip([i + ' ' + j for i in corralation.columns for j in corralation.columns],corralation.to_numpy().flatten())) #chuyen thanh dictionary
top = {key:val for key, val in top.items() if key.split() != key.split()[::-1] and abs(val) >= 0.95}
top = sorted(top.items(), key=lambda kv: kv[1]) #sort
```

```python
tbl = top[1::2]
```

```python
new_top = []
for i in tbl:
    a,b = i[0].split(' ')
    if symbols[(symbols.symbol == a)|(symbols.symbol == b)][['baseAsset','quoteAsset']].unstack().nunique() == 4:
        new_top += [i]        
```
